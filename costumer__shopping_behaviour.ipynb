```python
df = df.drop_duplicates()
```

```python
df.columns
```

```python
df = df.apply(lambda x : x.str.strip() if x.dtype == "object" else x)
#removes extra spaces
```

```python
#Standardize text cases
#gender column
df["Gender"] = df["Gender"].str.lower()
#Season column
df["Season"] = df["Season"].str.capitalize()
```

```python
#Handle currency symbols
#converts to numbers
df["Purchase Amount (USD)"] = (df["Purchase Amount (USD)"].astype(str).str.replace("$","",regex =False) )
df["Purchase Amout (USD)"] = pd.to_numeric( df["Purchase Amount (USD)"], errors = "coerce" )
```

```python
#Clean REview ratings
#remove text
df["Review Rating"] = (df["Review Rating"].astype(str).str.replace(" stars","",regex=False))
df["Review Rating"] = pd.to_numeric(df["Review Rating"],errors ="coerce")
```

```python
#Handle Invalid Age values
#checks invalid ages
df[df["Age"]<0]
#replace with NaN
df.loc[df["Age"]<0,"Age"] =np.nan
```

```python
#Fix MIxed Data Types
df["Previous Purchases"] = pd.to_numeric(df["Previous Purchases"],errors ="coerce")
```

```python
#Fill missing values
#numeric columns
numeric_cols =df.select_dtypes(include=np.number).columns
for col in numeric_cols:
  df[col] = df[col].fillna(df[col].median())
#categorical columns
cat_cols = df.select_dtypes(include ="object").columns
for col in cat_cols:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
```
